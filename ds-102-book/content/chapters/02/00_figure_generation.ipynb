```python
%matplotlib inline
import matplotlib.pyplot as plt

from daft import PGM
```

```python
m = PGM(dpi=100)
spacing = 1.7
offset = 0.8
m.add_node('a', r'$a$', -0.5 - 0.3, 1.9)
m.add_node('b', r'$b$', -0.5 + 0.3, 1.9)

for j, idx in [(-1.75*offset, '1'), (-0.75*offset, '2'), (offset*0.75, 'C')]:
    m.add_node(f'theta{idx}', rf'$\theta_{idx}$', j, 1)
    m.add_edge('a', f'theta{idx}')
    m.add_edge('b', f'theta{idx}')
    m.add_node(f'y{idx}', f'$y_{idx}$', j, 0.1, observed=True)
    m.add_node(f'n{idx}', f'$n_{idx}$', j-offset/2, 0.4, fixed=True)
    m.add_edge(f'theta{idx}', f'y{idx}')
    m.add_edge(f'n{idx}', f'y{idx}')
    m.add_text(-0.1, 1, r'$\ldots$')
    m.add_text(-0.1, 0.1, r'$\ldots$')
        

m.render()

plt.savefig('kc_hierarchical.png')
```

```python
m = PGM(dpi=100)
spacing = 1.7
offset = 0.5
m.add_node('ab', r'$a, b$', spacing/2, 1.5, fixed=True)
data = ['Axn', 'Bym']
for i, (ab, var, final_idx) in enumerate(data):
    i = i*spacing
    m.add_node(f'theta{ab}', rf'$\theta_{ab}$', i, 1)
    m.add_edge('ab', f'theta{ab}')
    var = 'x' if ab == 'A' else 'y'
    for j, idx in [(i-offset, '1'), (i+offset, final_idx)]:
        m.add_node(f'{var}{idx}', f'${var}_{idx}$', j, 0.1, observed=True)
        m.add_edge(f'theta{ab}', f'{var}{idx}')
    m.add_text(i-0.1, 0.1, r'$\ldots$')
        

m.render()

plt.savefig('review_model.png')
```

```python
m = PGM(dpi=100)
spacing = 1.7
offset = 0.5
m.add_node('mu', r'$\mu$', 0, 1)
m.add_node('x1', rf'$x_1$', -offset, 0)
m.add_node('x2', rf'$x_2$', offset, 0)
m.add_edge(f'mu', 'x1')
m.add_edge(f'mu', 'x2')
      

m.render()

plt.savefig('heights_gm.png')
```

```python
m = PGM(dpi=100)
spacing = 1.7
offset = 0.5
m.add_node('x', r'$x$', -1, 0)
m.add_node('y', '$y$', 0, 0)
m.add_node('z', '$z$', 1, 0)
m.add_edge('x', 'y')
m.add_edge('y', 'z')
      

m.render()

plt.savefig('chain_gm.png')
```
